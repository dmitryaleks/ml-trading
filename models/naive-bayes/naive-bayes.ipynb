# Naive Bayes model for classification

It is a classification technique based on Bayesâ€™ Theorem with an assumption of independence among predictors. In simple terms, a Naive Bayes classifier assumes that the presence of a particular feature in a class is unrelated to the presence of any other feature.

Bayes theorem provides a way of calculating posterior probability P(c|x) from P(c), P(x) and P(x|c):

![Bayes theorem](img/Bayse-theorem.png)

where:

  - P(c|x) is the posterior probability of class (c, target) given predictor (x, attributes).
  - P(c) is the prior probability of class.
  - P(x|c) is the likelihood which is the probability of predictor given class.
  - P(x) is the prior probability of predictor.

## Example using SciKit-Learn

```python
from sklearn.naive_bayes import GaussianNB
import numpy as np

# predictors
x= np.array([[-3,7],[1,5], [1,2], [-2,0], [2,3], [-4,0], [-1,1], [1,1], [-2,2], [2,7], [-4,1], [-2,7]])

# target variable
Y = np.array([3, 3, 3, 3, 5, 3, 3, 5, 3, 5, 5, 5])
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Extract individual features
x1 = [v[0] for v in x]
x2 = [v[1] for v in x]
```

```python
# Plot the target variable against a 2D feature space
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(np.ravel(x1), np.ravel(x2), Y, color='blue')
ax.set_title("Target variable against a 2D feature space", fontsize=16)
plt.show()
```

```python
# using Gaussian Naive Bayes classifier
model = GaussianNB()

model.fit(x, Y)

predictors = [0, 5]
# make a prediciton for a new unseen point 
predicted_class = model.predict([predictors])
print(f'Prediced class for features {predictors} is {predicted_class}')
```

```python
# Plot the target variable against a 2D feature space
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(np.ravel(x1), np.ravel(x2), Y, color='blue')
ax.scatter3D(np.ravel(predictors[0]), np.ravel(predictors[1]), predicted_class, color='green')
ax.set_title("Predicted class against a 2D feature space (in green)", fontsize=16)
plt.show()
```
